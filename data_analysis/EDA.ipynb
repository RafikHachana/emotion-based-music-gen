```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
import os
MUSIC_FEATURES = ['danceability', 'energy', 'key',
                  'loudness', 'mode', 'acousticness',
                  'instrumentalness', 'valence',
                  'tempo', 'time_signature']

if not os.path.exists("feature_dist"):
    os.makedirs("feature_dist")

for f in MUSIC_FEATURES:
    plt.figure()
    fig = sns.histplot(df[f], kde=True)
    plt.savefig(os.path.join("feature_dist", f + '.jpg'))
```

## Correlation

```python
plt.figure()
sns.heatmap(df[MUSIC_FEATURES].corr(), annot=True)
plt.savefig('correlation.jpg')
```

## Analyzing other features

Features that will not be used by the model but can give an idea on biases in the dataset.

```python
plt.figure()
sns.histplot(df['popularity'], kde=True)
plt.savefig(os.path.join("feature_dist", 'popularity.jpg'))
```

```python
import os
sns.set_theme(rc={'figure.figsize':(23,8.27)})
plt.figure()
df['track_genre'].value_counts()[:50].plot(kind='bar')
plt.savefig(os.path.join("feature_dist", 'track_genre.jpg'))
```
